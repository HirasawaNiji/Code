### 练习1.4
对元组做相同练习，并比较列表和元组的读取时间差异


````python
import random

# 生成一个随机的二维列表（100x100）
rows, cols = 100, 100
matrix = [[random.randint(1, 100) for _ in range(cols)] for _ in range(rows)]

# 方法一：外层循环遍历行，内层循环遍历列
def method1(matrix):
    for row in matrix:
        for item in row:
            pass  # 模拟打印，避免干扰性能分析

# 方法二：外层循环遍历列，内层循环遍历行
def method2(matrix):
    for col in range(len(matrix[0])):  # 假设列数相等
        for row in range(len(matrix)):
            pass  # 模拟打印，避免干扰性能分析

# 耗时分析
print("外层遍历行，内层遍历列（方法一）")
%timeit method1(matrix)
%time method1(matrix)

print("\n外层遍历列，内层遍历行（方法二）")
%timeit method2(matrix)
%time method2(matrix)
````

外层遍历行，内层遍历列（方法一）通常效率更高。

外层遍历列，内层遍历行（方法二）效率可能略低，尤其是当数据不连续存储（如大矩阵）时，内存访问可能不局部化。

````python
# 生成更大的二维列表和元组
rows, cols = 1000, 1000
large_list_matrix = [[random.randint(1, 100) for _ in range(cols)] for _ in range(rows)]
large_tuple_matrix = tuple(tuple(row) for row in large_list_matrix)

# 测试
print("大矩阵测试：二维列表")
%timeit method1(large_list_matrix)

print("\n大矩阵测试：二维元组")
%timeit method1(large_tuple_matrix)
````

对于大数据集，元组的性能优势更明显。

### 练习2.1 
①绘制三条不同相位和振幅的余弦函数，测试透视度、线条粗细、线条种类、颜色，用latex数学公式、annotate和legend标记。

````python
import numpy as np
import matplotlib.pyplot as plt

# 定义余弦函数参数
x = np.linspace(0, 2 * np.pi, 500)
y1 = 2 * np.cos(x)  # 振幅2，相位0
y2 = np.cos(x + np.pi / 4)  # 振幅1，相位π/4
y3 = 0.5 * np.cos(x - np.pi / 2)  # 振幅0.5，相位-π/2

# 创建绘图
plt.figure(figsize=(10, 6))

# 绘制三条余弦曲线
plt.plot(x, y1, label=r'$y=2\cos(x)$', color='blue', linestyle='-', linewidth=2, alpha=0.8)
plt.plot(x, y2, label=r'$y=\cos(x+\frac{\pi}{4})$', color='red', linestyle='--', linewidth=1.5, alpha=0.9)
plt.plot(x, y3, label=r'$y=0.5\cos(x-\frac{\pi}{2})$', color='green', linestyle='-.', linewidth=1, alpha=0.7)

# 添加注释
plt.annotate(r'$2\cos(x)$', xy=(np.pi/2, 2), xytext=(np.pi, 2.5),
             arrowprops=dict(facecolor='blue', arrowstyle='->'), fontsize=12)
plt.annotate(r'$\cos(x+\frac{\pi}{4})$', xy=(np.pi/4, 0), xytext=(1, -0.5),
             arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=12)
plt.annotate(r'$0.5\cos(x-\frac{\pi}{2})$', xy=(3*np.pi/2, -0.5), xytext=(4, -1.2),
             arrowprops=dict(facecolor='green', arrowstyle='->'), fontsize=12)

# 添加图例
plt.legend(fontsize=12)

# 设置标题和轴标签
plt.title("Cosine Functions with Different Phases and Amplitudes", fontsize=14)
plt.xlabel(r"$x$ (radians)", fontsize=12)
plt.ylabel(r"$y$", fontsize=12)

# 绘制shade背景
plt.fill_between(x, y3, color='green', alpha=0.2, label="Shaded Region ($0.5\cos$)")

# 设置网格
plt.grid(alpha=0.3)

# 显示图像
plt.show()
````

````python
# pip install memory_profiler
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import time
import tracemalloc

%load_ext memory_profiler

# ① 计算随机数的二维平均值和标准差，增加随机数的个数
def calculate_statistics(size):
    random_numbers = np.random.rand(size, size)  # 生成二维随机数
    mean = np.mean(random_numbers)
    std = np.std(random_numbers)
    return mean, std

# 初始测试
sizes = [10, 100, 1000]
for size in sizes:
    mean, std = calculate_statistics(size)
    print(f"Size: {size}x{size}, Mean: {mean:.4f}, Std: {std:.4f}")

# ② 利用直方图验证随机数是否均匀
random_uniform = np.random.rand(10000)
plt.figure(figsize=(10, 6))
plt.hist(random_uniform, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title("Histogram of Uniformly Distributed Random Numbers")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.show()

# ③ 利用直方图验证二维正态分布随机数是否符合正态分布
# 生成二维正态分布随机数
mean = [0, 0]  # 均值
cov = [[1, 0], [0, 1]]  # 协方差矩阵
normal_2d = np.random.multivariate_normal(mean, cov, size=10000)

# 构造三维直方图
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 使用直方图计算每个区域的频率
hist, xedges, yedges = np.histogram2d(normal_2d[:, 0], normal_2d[:, 1], bins=30, density=True)

# 构造网格坐标
xpos, ypos = np.meshgrid(xedges[:-1], yedges[:-1], indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = np.zeros_like(xpos)

# 构造柱状图的尺寸
dx = dy = (xedges[1] - xedges[0])  # 每个柱体的宽度
dz = hist.ravel()  # 每个柱体的高度

# 绘制三维直方图
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, shade=True, color='yellow')

# 设置图形标题和标签
ax.set_title("3D Histogram of Normally Distributed Random Numbers")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Density")

# 显示图形
plt.show()

# ④ 对随机数产生器做性能和内存分析
def performance_and_memory_analysis(size):
    tracemalloc.start()
    start_time = time.time()
    random_data = np.random.rand(size, size)
    elapsed_time = time.time() - start_time
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return elapsed_time, current / 1e6, peak / 1e6  # 返回时间和内存使用（MB）

# 测试不同规模的随机数生成
sizes = [100, 1000, 5000]
for size in sizes:
    elapsed_time, current_memory, peak_memory = performance_and_memory_analysis(size)
    print(f"Size: {size}x{size}, Time: {elapsed_time:.4f}s, Current Memory: {current_memory:.2f}MB, Peak Memory: {peak_memory:.2f}MB")

%prun calculate_statistics(10000) 
%memit calculate_statistics(10000)
````

### 练习4.2 用梯度下降法来计算函数$f(x)=x^4+2x^3-3x^2-2x$的局部和整体最小值

注意：设计算法计算，不可以人工分段
提示：可用修正梯度下降法

````python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数和导数
def f(x):
    return x**4 + 2*x**3 - 3*x**2 - 2*x

def f_prime(x):
    return 4*x**3 + 6*x**2 - 6*x - 2

# 梯度下降算法
def gradient_descent(f_prime, start_x, learning_rate=0.01, max_iter=1000, tolerance=1e-6):
    x = start_x
    for i in range(max_iter):
        gradient = f_prime(x)
        x_new = x - learning_rate * gradient
        
        # 检查收敛条件
        if abs(x_new - x) < tolerance:
            break
        
        x = x_new
    
    return x, f(x), i

# 设置初始点，找到局部和整体最小值
start_points = [-3, -1, 1, 3]
results = []

for start_x in start_points:
    min_x, min_f, iterations = gradient_descent(f_prime, start_x, learning_rate=0.01)
    results.append((min_x, min_f, iterations))

# 提取结果用于绘图
min_x_values = [result[0] for result in results]
min_y_values = [result[1] for result in results]

# 绘制函数和最小值点
x_values = np.linspace(-3.5, 3.5, 500)
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="f(x) = x^4 + 2x^3 - 3x^2 - 2x", color="blue")
plt.scatter(min_x_values, min_y_values, color="red", label="Local/Global Minima", zorder=5)
plt.title("Function and Minima Found by Gradient Descent")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()
plt.show()
````

### 练习5.3 计算方阵的逆矩阵linalg.inv()和转置.T，并验证是否单位矩阵相等eye()


````python
import numpy as np

# 定义一个方阵
matrix = np.array([[2, 1, 3],
                   [1, 3, 2],
                   [3, 2, 4]])

# 计算逆矩阵
inverse_matrix = np.linalg.inv(matrix)

# 计算转置矩阵
transpose_matrix = matrix.T

# 验证是否为单位矩阵 (matrix * inverse_matrix)
identity_matrix = np.dot(matrix, inverse_matrix)

# 创建单位矩阵
expected_identity = np.eye(matrix.shape[0])

# 打印结果
print("Original Matrix:")
print(matrix)

print("\nInverse Matrix:")
print(inverse_matrix)

print("\nTranspose Matrix:")
print(transpose_matrix)

print("\nMatrix * Inverse Matrix (Should be Identity):")
print(identity_matrix)

print("\nExpected Identity Matrix:")
print(expected_identity)

# 验证是否接近单位矩阵
is_identity = np.allclose(identity_matrix, expected_identity)
````


结果表明矩阵与其逆矩阵的乘积确实非常接近单位矩阵，验证成功

### 练习5.4 绘制向量与标准基底的图像

① 计算并绘制向量的基向量；
② 利用旋转矩阵将向量旋转逆时针45度角，并展示结果。


````python
import matplotlib.pyplot as plt

# 定义向量和标准基底
vector = np.array([2, 3])
basis_x = np.array([1, 0])  # 标准基底 x
basis_y = np.array([0, 1])  # 标准基底 y

# 定义旋转矩阵
theta = np.radians(45)  # 45度转为弧度
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

# 旋转向量
rotated_vector = np.dot(rotation_matrix, vector)

# 绘制向量和标准基底
plt.figure(figsize=(8, 8))

# 绘制标准基底
plt.quiver(0, 0, basis_x[0], basis_x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Basis X')
plt.quiver(0, 0, basis_y[0], basis_y[1], angles='xy', scale_units='xy', scale=1, color='green', label='Basis Y')

# 绘制原始向量
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='red', label='Original Vector')

# 绘制旋转后的向量
plt.quiver(0, 0, rotated_vector[0], rotated_vector[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Rotated Vector (45°)')

# 设置图形范围
plt.xlim(-4, 4)
plt.ylim(-4, 4)

# 添加网格和图例
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.grid()
plt.legend()
plt.title("Vectors and Basis with 45° Rotation")
plt.xlabel("x")
plt.ylabel("y")

# 显示图像
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

````

### 7.2 性能分析``%prun``

````python
def sum_of_lists(N):
    total = 0
    for i in range(5):
        L = [j ^ (j >> i) for j in range(N)]
        total += sum(L)
    return total
````

````python
%prun sum_of_lists(1000000)
````

结果显示了每个函数所耗费的资源。
